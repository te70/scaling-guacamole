##### Effects on Borrower's APR

##### Investigation overview

> In this investigation, we look at the characteristics of loans that could be used to predict their borrower APR. The main focus was on the original loan amount, borrower's stated monthly income, , loan term, borrower's Prosper rating.

##### Dataset overview
> The dataset consisted of borrower APRs and attributes of 113,937 loans. The attributes included original loan amount, borrower's Prosper rating, borrower's stated monthly income, loan term, as well as many other features such as borrower's employment status, current loan status, debt to income ratio. Some data points were removed from the analysis due to very large stated monthly income seemed as outliers and missing borrower APR information.

```python
#import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
#load in the dataset into a pandas dataframe
df = pd.read_csv('prosperLoanData.csv')
#dataframe set by selecting features of interest
cols = ['BorrowerAPR','LoanOriginalAmount', 'ProsperRating (Alpha)', 'StatedMonthlyIncome', 'Term',  'EmploymentStatus']

df_set = df[cols]

#remove loans with missing borrower APR information
df_set = df_set[~df_set.BorrowerAPR.isna()]

#remove loans with stated monthly income greater than 30000
df_set = df_set[df_set.StatedMonthlyIncome<=30000]

#convert prosperrating and employment stauts into ordered categorical types
rate_order = ['HR', 'E', 'D', 'C', 'B', 'A', 'AA']
or_var = pd.api.types.CategoricalDtype(ordered = True, categories = rate_order)
df_set['ProsperRating (Alpha)'] = df_set['ProsperRating (Alpha)'].astype(or_var)

er_order = ['Employed', 'Self-employed', 'Full-time', 'Part-time', 'Retired', 'Other', 'Not employed', 'Not available']
or_var = pd.api.types.CategoricalDtype(ordered = True, categories = er_order)
df_set['EmploymentStatus'] = df_set['EmploymentStatus'].astype(or_var)
df_set.shape
```

##### Distribution of Borrower APR
> The distribution of Borrower APR seems to be multimodal. A small peak at 0.1, a large peak centered at 0.2. There is also a small peak centered at 0.3. Only very few loans have APR greater than 0.4.

```python
bins = np.arange(0, df_set.BorrowerAPR.max()+0.05, 0.01)
plt.figure(figsize=[8,5])
plt.hist(data = df_set, x='BorrowerAPR', bins = bins)
plt.xlabel('Borrower APR')
plt.title('Borrower APR')
```

##### Distribution of original loan amount
> The large spikes in frequenct are at 10000, 15000, 2000, 25000 and 35000.

```python
bins = np.arange(8000, df_set.LoanOriginalAmount.max()+200, 200)
plt.figure(figsize=[8,5])
plt.hist(data = df_set, x = 'LoanOriginalAmount', bins = bins)
plt.xlabel('Loan original amount')
plt.title('Distribution of loan original amount')
```

##### Borrower APR vs Loan Amount
> At different amount of the loan amount, but the range of APR decrease witht the increase of loan amount. Overall, the borower APR decrease with increase of loan amount

```python
plt.figure(figsize=[8,6])
sb.regplot(data = df_set, x='LoanOriginalAmount', y='BorrowerAPR')
plt.xlabel('Loan Amount')
plt.ylabel('Borrower APR')
plt.title('Borrower APR vs. Loan Amount')
```

##### Borrower APR vs Prosper rating
> The borrower APR decreases with the increasingly better rating. Borrowers with the best prosper ratings have the lowest APR. IProsper rating and borrower APR has strong correlation

```python
plt.figure(figsize=[9,6])
sb.scatterplot(data = df_set, x = 'ProsperRating (Alpha)', y='BorrowerAPR')
plt.xlabel('Prosper Rating')
plt.ylabel('Borrower APR')
plt.title('Borrower APR vs Prosper Rating')
```

##### Prosper rating effect on APR and loan amount
> The borrower APR decreases with better rating. The loan amount increases with better rating. Interestingly, the relationship between borrower APR and loan amount turns from negative to slightly positive when the prosper ratings are increases from HR to A or better. 

```python
f = sb.FacetGrid(data = df_set, col = 'ProsperRating (Alpha)', height=3.5, col_wrap=4)
f.map(sb.scatterplot, 'LoanOriginalAmount', 'BorrowerAPR')
f.set_titles('{col_name}')
f.add_legend();
f.set_xlabels('Loan Amount')
f.set_ylabels('Borrower APR')
plt.suptitle('Borrower APR vs loan amount')
plt.subplots_adjust(top=0.85)
```

##### Borrower APR by rating and term
The borrower APR decrease with the increase of borrow term for people with HR-C ratings. But for people with B-AA ratings, the APR increase with the increase of borrow term. 

```python
fig = plt.figure(figsize = [8,6])
ax = sb.boxplot(data = df_set, x = 'ProsperRating (Alpha)', y='BorrowerAPR')
plt.title('Borrower APR across rating and term')
plt.ylabel('Mean Borrower APR')
plt.xlabel('Prosper Rating')
plt.title('Borrower APR by rating and term')
```
